## Load and Combine Data

### Subtask:
Load all 'api_data_aadhar_demographic_*.csv' files into a single pandas DataFrame to consolidate the entire dataset for analysis.


**Reasoning**:
First, I will import the necessary libraries, pandas and glob, which are essential for file operations and data manipulation. Then, I will find all the CSV files matching the specified pattern, read them one by one into pandas DataFrames, concatenate them into a single DataFrame, and finally display its head and summary to verify the data loading and combination.



```python
import pandas as pd
import glob

# Find all files matching the pattern
file_paths = glob.glob('api_data_aadhar_demographic_*.csv')

# Initialize an empty list to hold individual DataFrames
df_list = []

# Loop through each file path and read the CSV into a DataFrame
for file in file_paths:
    df = pd.read_csv(file)
    df_list.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
df_combined = pd.concat(df_list, ignore_index=True)

# Display the first 5 rows of the df_combined DataFrame
print("First 5 rows of the combined DataFrame:")
print(df_combined.head())

# Print the concise summary of the df_combined DataFrame
print("\nConcise summary of the combined DataFrame:")
df_combined.info()
```

## Initial Data Exploration

### Subtask:
Perform an initial exploration of the combined dataset, including checking data types, identifying missing values, and generating summary statistics to understand the data's structure and quality.


**Reasoning**:
The previous step already performed `df_combined.info()` to check data types and non-null values. Now, to continue the initial data exploration, I will generate descriptive statistics for both numerical and non-numerical columns of the `df_combined` DataFrame to understand their distribution and unique values, respectively.



```python
print("\nDescriptive statistics for numerical columns:")
print(df_combined.describe())

print("\nDescriptive statistics for non-numerical (object) columns:")
print(df_combined.describe(include='object'))
```

## Demographic Distribution Analysis

### Subtask:
Analyze the distribution of key demographic features such as age groups, states, and districts to identify patterns and concentrations within the dataset.


**Reasoning**:
To analyze the distribution of age groups, states, and districts, I will calculate the sum of individuals in each age group ('demo_age_5_17' and 'demo_age_17_') and then find the top 10 most frequent states and districts using value counts.



```python
print("\nTotal individuals in 'demo_age_5_17' age group:")
print(df_combined['demo_age_5_17'].sum())

print("\nTotal individuals in 'demo_age_17_' age group:")
print(df_combined['demo_age_17_'].sum())

print("\nTop 10 most frequent states:")
print(df_combined['state'].value_counts().head(10))

print("\nTop 10 most frequent districts:")
print(df_combined['district'].value_counts().head(10))
```

## Visualize Key Demographics

### Subtask:
Create appropriate visualizations (e.g., bar charts, histograms) to represent the distributions of the most significant demographic features, making sure to include legends.


**Reasoning**:
To visualize the distribution of demographic features, I will import the necessary plotting libraries, `matplotlib.pyplot` and `seaborn`, then create bar charts for age groups, top 10 states, and top 10 districts as specified in the instructions, and finally display them.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Bar chart for age group distribution
age_groups = ['demo_age_5_17', 'demo_age_17_']
age_counts = [df_combined['demo_age_5_17'].sum(), df_combined['demo_age_17_'].sum()]

plt.figure(figsize=(8, 6))
sns.barplot(x=age_groups, y=age_counts, palette='viridis')
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Total Individuals')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 3. Bar chart for top 10 most frequent states
top_10_states = df_combined['state'].value_counts().head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_states.index, y=top_10_states.values, palette='magma')
plt.title('Top 10 Most Frequent States')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 4. Bar chart for top 10 most frequent districts
top_10_districts = df_combined['district'].value_counts().head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_districts.index, y=top_10_districts.values, palette='plasma')
plt.title('Top 10 Most Frequent Districts')
plt.xlabel('District')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous code produced `FutureWarning` messages from Seaborn due to a change in how `palette` is handled. To fix this, I need to explicitly assign the `x` variable to `hue` and set `legend=False` for each `sns.barplot` call to align with the recommended usage in newer Seaborn versions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Bar chart for age group distribution
age_groups = ['demo_age_5_17', 'demo_age_17_']
age_counts = [df_combined['demo_age_5_17'].sum(), df_combined['demo_age_17_'].sum()]

plt.figure(figsize=(8, 6))
sns.barplot(x=age_groups, y=age_counts, hue=age_groups, palette='viridis', legend=False)
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Total Individuals')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 3. Bar chart for top 10 most frequent states
top_10_states = df_combined['state'].value_counts().head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_states.index, y=top_10_states.values, hue=top_10_states.index, palette='magma', legend=False)
plt.title('Top 10 Most Frequent States')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 4. Bar chart for top 10 most frequent districts
top_10_districts = df_combined['district'].value_counts().head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_districts.index, y=top_10_districts.values, hue=top_10_districts.index, palette='plasma', legend=False)
plt.title('Top 10 Most Frequent Districts')
plt.xlabel('District')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
